```python
import pandas as pd
```

```python
#Read in the e-commerce purchases data which is in a csv format
Ecommerce = pd.read_csv('C:\data\Ecommerce_purchases.csv')
```

```python
#Checking the head of the Dataset.
Ecommerce.head()
```

```python
#Get the Info
Ecommerce.info()
```

```python
#Get the average purchase price
mean=Ecommerce['Purchase Price'].mean()
mean
```

```python
#Get the Minimum and Maximum purchase price
Ecommerce[['Purchase Price']].agg(['min','max'])
##print ("minimum purchase price =",)
##print ("maximum purchase price =",)
```

```python
#obtaining the number of people subscribed to english which is 'en'.
Ecommerce.loc[Ecommerce['Language'] =='en']
```

```python
#Obtaining the number of people with job title of "Lawyer" ?
Ecommerce.loc[Ecommerce['Job'] =='Lawyer'].head()
```

```python
#Getting the number of people who made purchase during the AM and people made the purchase during PM 
Ecommerce.loc[Ecommerce["AM or PM"] == "AM", 'Purchase Price' ], Ecommerce.loc[Ecommerce["AM or PM"] == "AM", 'Purchase Price' ]
```

```python
#Getting the 5 most common Job Titles?
Ecommerce = Ecommerce.sort_values('Job', ascending=False).head(5)
Ecommerce
```

```python
#Getting the purchase that came from Lot: "90 WT" , and the Purchase Price for the transaction.
Ecommerce.loc[Ecommerce["Lot"] == "90 WT", 'Purchase Price' ]
```

```python
#Getting the email of the person with the following Credit Card Number: 4926535242672853

Ecommerce.loc[Ecommerce["Credit Card"] == 4926535242672853, 'Email']
```

```python
#No. of people who have American Express as their Credit Card Provider and made a purchase above $95

len(Ecommerce[(Ecommerce['CC Provider']== "American Express") & ["Purchase Price"]>95])
```

```python
#Getting the number of people who have a credit card that expires in 2025

sum(Ecommerce["CC Exp Date"].apply(lambda exp: exp[3:]=='25'))
```

```python
#Getting the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...)

Ecommerce['Email'].apply(lambda email: email. split('@')[1]).value_counts().head()
```
